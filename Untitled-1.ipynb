```python
def resize_image(image, target_depth):
    """Resize the depth dimension of the image to the target depth."""
    depth_factor = target_depth / image.shape[2]
    resized_image = zoom(image, (1, 1, depth_factor), order=1)  # Linear interpolation
    return resized_image
```

```python
import nibabel as nib
import numpy as np
import os
from scipy.ndimage import zoom


def load_nifti_image(path):
    """Load a NIfTI image and return it as a numpy array."""
    img = nib.load(path)
    img_data = img.get_fdata()  # (x, y, z)
    return img_data

def normalize_image(image):
    """Normalize the image to the range [0, 1] after clipping to the typical CT range."""
    image = np.clip(image, -200, 250)  # Clip to typical CT range
    image = (image - np.min(image)) / (np.max(image) - np.min(image))
    return image

# Example paths
train_path = r"C:\Users\ishan\Downloads\FLARE22Train\FLARE22Train\images"
label_path = r"C:\Users\ishan\Downloads\FLARE22Train\FLARE22Train\labels"

num_samples = 5

# Load a small subset of training data and labels
train_files = sorted(os.listdir(train_path))[:num_samples]
label_files = sorted(os.listdir(label_path))[:num_samples]

# Ensure that the number of images and labels match
assert len(train_files) == len(label_files), "Mismatch between number of images and labels."

# Define target depth for resizing
target_depth = 80  # Example depth, adjust as needed

# Load and process images and labels
# Load and process images and labels
train_images = np.array([resize_image(normalize_image(load_nifti_image(os.path.join(train_path, f))), target_depth) for f in train_files])
labels = np.array([resize_image(load_nifti_image(os.path.join(label_path, f)), target_depth) for f in label_files])


# Add channel dimension if needed (e.g., for models expecting a channel dimension)
train_images = np.expand_dims(train_images, axis=-1)  # shape (num_samples, height, width, depth, channels)
labels = np.expand_dims(labels, axis=-1)  # shape (num_samples, height, width, depth, channels)

# Check shapes to confirm
print(f"Train images shape: {train_images.shape}")
print(f"Labels shape: {labels.shape}")
```

```python
def print_image_shapes(path):
    """Print the shape of each NIfTI image in the specified directory."""
    files = [f for f in os.listdir(path) if f.endswith('.nii.gz')]
    
    for file in files:
        img_path = os.path.join(path, file)
        img = nib.load(img_path)
        img_data = img.get_fdata()
        print(f"File: {file}, Shape: {img_data.shape}")

# Check shapes of training images
print("Training Images:")
print_image_shapes(train_path)
```

```python
def resample_image(image, target_shape):
    """Resample the image to the target shape."""
    factors = np.array(target_shape) / np.array(image.shape)
    resampled_image = zoom(image, factors, order=1)  # order=1 is linear interpolation
    return resampled_image

def resample_and_save_images(input_path, output_path, target_shape):
    """Resample all images in the directory to the target shape and save them."""
    files = [f for f in os.listdir(input_path) if f.endswith('.nii.gz')]
    
    for file in files:
        img_path = os.path.join(input_path, file)
        img = nib.load(img_path)
        img_data = img.get_fdata()
        
        # Resample the image
        resampled_data = resample_image(img_data, target_shape)
        
        # Create a new NIfTI image
        resampled_img = nib.Nifti1Image(resampled_data, img.affine)
        
        # Save the resampled image
        output_file_path = os.path.join(output_path, file)
        nib.save(resampled_img, output_file_path)
        
        print(f"File: {file}, Original Shape: {img_data.shape}, Resampled Shape: {resampled_data.shape}")

# Example usage
input_train_path = r"C:\Users\ishan\Downloads\FLARE22Train\FLARE22Train\images"
output_train_path = r"C:\Users\ishan\Downloads\FLARE22Train\FLARE22Train\resampled_images"
target_shape = (256, 256, 80)

# Ensure the output directory exists
os.makedirs(output_train_path, exist_ok=True)

# Resample and save images
resample_and_save_images(input_train_path, output_train_path, target_shape)
```

```python
def resample_labels(input_path, output_path, target_shape):
    """Resample all label images in the directory to the target shape."""
    files = [f for f in os.listdir(input_path) if f.endswith('.nii.gz')]
    
    if not os.path.exists(output_path):
        os.makedirs(output_path)

    for file in files:
        label_path = os.path.join(input_path, file)
        label_img = nib.load(label_path)
        label_data = label_img.get_fdata()

        # Resample the label image
        resampled_label_data = resample_image(label_data, target_shape)

        # Save the resampled label
        resampled_label_img = nib.Nifti1Image(resampled_label_data, label_img.affine)
        resampled_label_path = os.path.join(output_path, file)
        nib.save(resampled_label_img, resampled_label_path)

        print(f"File: {file}, Original Shape: {label_data.shape}, Resampled Shape: {resampled_label_data.shape}")

# Example usage
input_label_path = r"C:\Users\ishan\Downloads\FLARE22Train\FLARE22Train\labels"
output_label_path = r"C:\Users\ishan\Downloads\FLARE22Train\FLARE22Train\resampled_labels"
target_shape = (256, 256, 80)  # Match this with the resampled image shape

resample_labels(input_label_path, output_label_path, target_shape)
```
